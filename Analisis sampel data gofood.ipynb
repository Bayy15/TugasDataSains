```python
import pandas as pd
import math as mt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/Rtech/Downloads/food_delivery_datasets.csv")
data = df.drop(columns=["random_id",'date_time','eta_seconds'])
data_gofood=data.fillna({'rating':0})
data_gofood
```

```python
dcc=data_gofood.loc[(data_gofood.status == 'Completed')&(data_gofood.duplicated('cust_id'))]
dcc
```

```python
dcc_b5=dcc.loc[(dcc.rating == 5.0)]
dcc_b5
```

```python
all_cust_comp=len(dcc)
all_cust_comp
```

```python
all_cust_b5=len(dcc_b5)
all_cust_b5
```

```python
all_cust_norate=len(dcc.loc[dcc.rating == 0.0])
all_cust_norate
```

```python
lainnya=all_cust_comp-(all_cust_b5+all_cust_norate)
lainnya
```

```python
labels = "Rating bintang 5", "No Rating", "Lainnya"
sizes = [all_cust_b5,all_cust_norate,lainnya]

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  
plt.show()
```
